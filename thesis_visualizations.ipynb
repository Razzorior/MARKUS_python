```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# PLOT FOR BACKGROUND / Over- AND UNDERFITTING

# Sample data for demonstration
epochs = np.arange(1, 21)
train_acc_overfit = 1 - np.exp(-0.1 * epochs)
val_acc_overfit = 1 - np.exp(-0.1 * epochs) + (np.exp(-0.03 * epochs) - 1)
#val_acc_overfit[10:] -= 0.1  # Simulate overfitting by decreasing validation accuracy

train_acc_underfit = (1 - np.exp(-0.1 * epochs)) * 0.4
val_acc_underfit = (1 - np.exp(-0.07 * epochs)) * 0.4

train_acc_proper = 1 - np.exp(-0.1 * epochs)
val_acc_proper = 1 - np.exp(-0.085 * epochs)

plt.rcParams.update({'font.size': 20})

sns.set_style("whitegrid", {'axes.facecolor': 'none', 'figure.facecolor': 'none'})

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

fig.patch.set_facecolor('white')

sns.lineplot(ax=axes[1], x=epochs, y=train_acc_overfit, label='Training Accuracy', legend=False)
sns.lineplot(ax=axes[1], x=epochs, y=val_acc_overfit, label='Validation Accuracy', legend=False)
axes[1].set_title('Overfitting')
axes[1].set_xlabel('Epochs')
axes[1].set_ylabel('Accuracy')
axes[1].tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
axes[1].grid(False)
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].spines['left'].set_visible(True)
axes[1].spines['bottom'].set_visible(True)

axes[1].annotate('', xy=(1, 0), xytext=(0, 0),
                 arrowprops=dict(arrowstyle="->", color='black', lw=1),
                 xycoords='axes fraction', textcoords='axes fraction')
axes[1].annotate('', xy=(0, 1), xytext=(0, 0),
                 arrowprops=dict(arrowstyle="->", color='black', lw=1),
                 xycoords='axes fraction', textcoords='axes fraction')

# Underfitting plot
sns.lineplot(ax=axes[0], x=epochs, y=train_acc_underfit, label='Training Accuracy', legend=False)
sns.lineplot(ax=axes[0], x=epochs, y=val_acc_underfit, label='Validation Accuracy', legend=False)
axes[0].set_title('Underfitting')
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Accuracy')
axes[0].tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
axes[0].grid(False)
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)
axes[0].spines['left'].set_visible(True)
axes[0].spines['bottom'].set_visible(True)


axes[0].annotate('', xy=(1, 0), xytext=(0, 0),
                 arrowprops=dict(arrowstyle="->", color='black', lw=1),
                 xycoords='axes fraction', textcoords='axes fraction')
axes[0].annotate('', xy=(0, 1), xytext=(0, 0),
                 arrowprops=dict(arrowstyle="->", color='black', lw=1),
                 xycoords='axes fraction', textcoords='axes fraction')



# Properly trained model plot
sns.lineplot(ax=axes[2], x=epochs, y=train_acc_proper, label='Training Accuracy', legend=False)
sns.lineplot(ax=axes[2], x=epochs, y=val_acc_proper, label='Validation Accuracy', legend=False)
axes[2].set_title('Well generalized model')
axes[2].set_xlabel('Epochs')
axes[2].set_ylabel('Accuracy')
axes[2].tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
axes[2].grid(False)
axes[2].spines['top'].set_visible(False)
axes[2].spines['right'].set_visible(False)
axes[2].spines['left'].set_visible(True)
axes[2].spines['bottom'].set_visible(True)


axes[2].annotate('', xy=(1, 0), xytext=(0, 0),
                 arrowprops=dict(arrowstyle="->", color='black', lw=1),
                 xycoords='axes fraction', textcoords='axes fraction')
axes[2].annotate('', xy=(0, 1), xytext=(0, 0),
                 arrowprops=dict(arrowstyle="->", color='black', lw=1),
                 xycoords='axes fraction', textcoords='axes fraction')


# Create a custom legend
lines_labels = [axes[0].get_lines(), ['Training accuracy', 'Validation accuracy']]
lines, labels = lines_labels
fig.legend(lines, labels, loc='lower center',bbox_to_anchor=(0.5, -0.12), ncol=2)

plt.tight_layout()
plt.show()
```

```python
# PLOT FOR BACKGROUND / EXPLODING- & VANISHING GRADIENTS
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate data for exploding gradients
iterations = np.arange(1, 101)
exploding_gradients = np.exp(0.1 * iterations)

# Generate data for vanishing gradients
vanishing_gradients = np.exp(-0.1 * iterations)

# Set up the plots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot exploding gradients
sns.lineplot(x=iterations, y=exploding_gradients, ax=axes[0])
axes[0].set_title('Exploding Gradients')
axes[0].set_xlabel('Training Iterations')
axes[0].set_ylabel('Gradient Values')

# Plot vanishing gradients
sns.lineplot(x=iterations, y=vanishing_gradients, ax=axes[1])
axes[1].set_title('Vanishing Gradients')
axes[1].set_xlabel('Training Iterations')
axes[1].set_ylabel('Gradient Values')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the activation functions
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

plt.rcParams.update({'font.size': 20})
sns.set_style("whitegrid", {'axes.facecolor': 'none', 'figure.facecolor': 'none'})

# Generate x values
x = np.linspace(-3, 3, 400)
x1 = np.linspace(-9, 9, 400)

# Generate y values for each activation function
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot sigmoid
sns.lineplot(x=x1, y=y_sigmoid, ax=axes[0])
axes[0].set_title('Sigmoid')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sigmoid(x)')

# Plot tanh
sns.lineplot(x=x, y=y_tanh, ax=axes[1])
axes[1].set_title('Tanh')
axes[1].set_xlabel('x')
axes[1].set_ylabel('tanh(x)')

# Plot ReLU
sns.lineplot(x=x, y=y_relu, ax=axes[2])
axes[2].set_title('ReLU')
axes[2].set_xlabel('x')
axes[2].set_ylabel('ReLU(x)')

# Show the plots
plt.tight_layout()
plt.show()
```
